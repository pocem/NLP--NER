```python
import os
import sys
from medmentions import MedMentionsDataset

# 1. Manually create the builder instance with your desired config
builder = MedMentionsDataset(config_name="medmentions_full_source")

# 2. Download and prepare the data 
# (This will handle the URLs defined in your _URLS dictionary)
builder.download_and_prepare()

# 3. Load it as a standard Hugging Face Dataset object
dataset = builder.as_dataset()

# 4. Check the results
print(dataset)
print(dataset['train'][0])
```
